```python
#W9A1
def merge(lst,arr): 
    m=[]
    i,j =0,0 
    while i<len(lst) and j< len(arr): 
        if lst[i]<=arr[j]: 
            m.append(lst[i])
            i+=1
        else: 
            m.append(arr[j])
            j+=1
    m.extend(lst[i:])
    m.extend(arr[j:])
    return m 
lst = list(map(int,input().split()))
arr= list(map(int,input().split()))
print(merge(lst,arr))
```

```python
#W9A3 
def t(n):
    s = list(str(n))
    length = len(s)

    for i in range(length):
        min_digit = s[i]
        pos = -1

        for j in range(i + 1, length):
            if s[j] < min_digit:
                min_digit = s[j]
                pos = j

        # tránh tạo số có 0 ở đầu
        if pos != -1 and not (i == 0 and s[pos] == '0'):
            s[i], s[pos] = s[pos], s[i]
            break

    return int("".join(s))
n = int(input())
print(t(n))

```

```python
#W9A4
def canTransform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    for i in range(1, n + 1):
        for j in range(m + 1):
            if a[i - 1].islower():
                if dp[i - 1][j]:
                    dp[i][j] = True
                if j > 0 and a[i - 1].upper() == b[j - 1] and dp[i - 1][j - 1]:
                    dp[i][j] = True
            else:
                if j > 0 and a[i - 1] == b[j - 1] and dp[i - 1][j - 1]:
                    dp[i][j] = True

    return dp[n][m]
Q   = int(input())
a,b = map(str,input().split())
t = canTransform(a, b)
if t : 
    print("YES") 
else : 
    print("NO")
```

```python
#W9A5
def kiem_tra(list1, list2):
    n = len(list1)
    m = len(list2)

    if n > m:
        return False
    for i in range(m - n + 1):
        if list2[i:i+n] == list1:
            return True
    return False
list1 = list(map(int,input().split()))
list2 = list(map(int,input().split()))
t=kiem_tra(list1,list2)
if t : 
    print("YES" )
else : 
    print("NO") 
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)

    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i - 1] == 0)
            right = (i == n - 1) or (flowerbed[i + 1] == 0)

            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
    return k <= 0
flowerbed= list(map(int,input().split()))
k= int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A9 
def Median(nums):
    sum=0
    for i in nums:
        sum+=i
    return sum//5    
nums=list(map(int,input().split()))
if len(nums)!=5:
    nums=list(map(int,input().split()))
print(Median(nums))
```

```python
#W9A10 
def findRadius(houses,heaters):
  houses.sort()
  heaters.sort()
  radius=0
  i=0
  for house in houses:
    while i<len(heaters) -1 and abs(heaters[i+1]-house)<=abs(heaters[i]-house):
      i+=1
    curr=abs(heaters[i]-house)
    radius=max(radius,curr)
  return radius
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
print(findRadius(houses,heaters))
```
