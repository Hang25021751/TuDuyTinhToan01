```python
#W8A1 
class Cylinder(): 
    def __init__(self,ban_kinh,chieu_cao):
        self.chieu_cao=chieu_cao 
        self.ban_kinh=ban_kinh
    def Dien_tich_toan_phan(self):
        S=2*3.14*(self.ban_kinh)*(self.ban_kinh+self.chieu_cao)
        return f"{S:.2f}"
    def The_tich(self): 
        V=3.14*(self.ban_kinh)**2*(self.chieu_cao)
        return f"{V:.2f}"
bankinh,chieucao=map(float,input().split())
Hinhtru= Cylinder(bankinh,chieucao)
print(Hinhtru.Dien_tich_toan_phan())
print(Hinhtru.The_tich())
```

```python
#W8A2 
class Date(): 
    def __init__(self,day, month,year): 
        self.day = day 
        self.month= month
        self.year= year 
    def kiem_tra(self):
        if self.day < 1: 
            return False 
        if  self.month<1 or self.month > 12: 
            return False 
        if self.month in (4,6,9,11): 
            if self.day > 30 :
                return False 
        if self.month in (1,3,5,7,8,10,12): 
            if self.day > 31: 
                return False 
        if (self.year%4==0 and self.year%100!=0) or (self.year%400==0): 
            if self.month==2: 
                if self.day > 29: 
                    return False 
        else:
            if self.month==2: 
                if self.day > 28 : 
                    return False 
        return True 
    def ngay_ke_tiep(self): 
        if self.month in (4,6,9,11): 
            if self.day ==30:
                self.day=1
                self.month+=1
            else:
                self.day +=1
        elif self.month in (1,3,5,7,8,10,12): 
            if self.day==31: 
                self.day=1
                if self.month ==12: 
                    self.month=1
                    self.year+=1
                else:
                    self.month+=1
            else:
                self.day+=1
        elif  self.month==2:
            if (self.year%4==0 and self.year%100!=0) or (self.year%400==0): 
                if self.day == 29: 
                    self.day = 1
                    self.month+=1
                else:
                    self.day+=1
            else:
                if self.day==28:
                    self.day=1
                    self.month+=1
                else:
                    self.day+=1
        return f"{self.day:02d}/{self.month:02d}/{self.year}" 
day,month,year= map(int,input().split("/"))
D= Date(day,month,year)
if  not D.kiem_tra(): 
    print("INVALID")
else:
    print(D.ngay_ke_tiep())
                
```

```python
#W8A3 
class Diem(): 
    def __init__(self,toa_do,tung_do): 
        self.x= toa_do
        self.y=tung_do
    def kiem_tra(self): 
        if self.x==0 and self.y!=0:
            return ("diem thuoc truc hoanh")
        elif self.x!=0 and self.y==0: 
            return("diem thuoc truc tung")
        elif self.x==0 and self.y==0: 
            return ("diem la goc toa do")
        else: 
            return(" diem khong thuoc truc hoanh,truc tung,khong la goc toa do")
    def khoang_cach(self): 
        d = ((self.x)**2+(self.y)**2)**0.5
        return f"{d:.2f}" 
x,y = map(float,input().split())
M = Diem(x,y)
print(M.kiem_tra())
print(M.khoang_cach())

```

```python
#W8A6 
class Rectangle():
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return (f"{self.w*self.h:.2f}")
    def perimeter(self):
        return(f"{2*(self.w+self.h):.2f}")
    def scale(self,k):
        self.w*=k
        self.h*=k
w,h,k= map(float,input().split())
Hinhchunhat=Rectangle(w,h)
Hinhchunhat.scale(k)
print(Hinhchunhat.area())
print(Hinhchunhat.perimeter())
```

```python
#W8A8 
class BankAccount():
    def __init__(self,owner,balance):
        self.owner= owner
        self.balance= balance
    def deposit(self,x):
        self.balance+=x
        return f"{self.balance:.2f}"
    def withdraw(self,x):
        self.balance-=x
        return f"{self.balance:.2f}"
owner= input("nhập vào chủ tài khoản")
balance=float(input("số dư ban đầu"))
Bank= BankAccount(owner,balance)
n = int(input())
for i in range(n):
    x = float(input("biến động số dư"))
    choice=input("Rút hay gửi ")
    if choice=="DEPOSIT":
        Bank.deposit(x)
    elif choice=="WITHDRAW":
        if Bank.balance < x:
            continue
        else:
            Bank.withdraw(x)
print(f"{Bank.balance:.2f}")


```

```python
#W8A9 
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}   # lưu {subject: score}

    def add_score(self, subject, score):
        self.scores[subject] = score

    def avg(self):
        if not self.scores:
            return 0
        return sum(self.scores.values()) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"
name = input("nhap vao ten") 
S= Student(name) 
n = int(input()) 
for i in range(n): 
    subject= input(" Nhap vao mon hoc") 
    score= float(input("nhap vao diem"))
    S.add_score(subject,score)
print(f"{S.name} {S.avg():.2f} {S.rank()}")
```

```python
#W8A10 
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, year):
        return sum(1 for b in self.books if b.year == year)
```

```python
#W8A12
class Complex:
    def __init__(self, realPart, imaginaryPart):
        self.real = realPart
        self.imag = imaginaryPart

    def multiply(self, other):
        real_new = self.real * other.real - self.imag * other.imag
        imag_new = self.real * other.imag + self.imag * other.real
        return Complex(real_new, imag_new)
a, b = map(int, input().split())
c, d = map(int, input().split())
z1 = Complex(a, b)
z2 = Complex(c, d)
result = z1.multiply(z2)
print(result.real, result.imag)

```
