```python
#W6A9
m,n= map(int,input().split())
dic1 ,dic2 ,dic3={},{},{}
for i in range(n):
    k = input()
    v = int(input())
    dic1[k]=v 
for j in range(m):
    q = input()
    h = int(input())
    dic2[q]=h 
for k,v in dic1.items():
    if k in dic2:
        dic3[k]=v+dic2[k]
    else:
        dic3[k]=v 
for q,h in dic2.items():
    if q not in dic3:
        dic3[q]=h 
l=dict(sorted(dic3.items()))
print(l)
```
