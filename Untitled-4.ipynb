```python
#W5A1 
def so_sanh(a,b):
    if a>=b:
        return a 
    else:
        return b 
a,b = map(int,input().split())
print(so_sanh(a,b))
```

```python
#W5A2 
def swap(a,b): 
    t=a
    a=b
    b = t
    return a,b 
a,b = map(int,input().split())
print(swap(a,b))
```

```python
#W5A3 
def so_nguyen_to(n):
    if n<=1:
        return False 
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False 
    return True 
n = int(input())
print(so_nguyen_to(n))
```

```python
#W5A4
def so_hoan_hao(n):
    t =0
    for i in range(1,n):
        if n%i==0:
            t+=i
    if t==n:
        return True 
    return False 
n = int(input())
print(so_hoan_hao(n))
```

```python
#W5A5 
def t(arr,k):
    for i in range(len(arr)):
        if arr[i]==k:
            return i+1
    return -1
arr = list(map(int,input().split()))
k = int(input())
print(t(arr,k))
```

```python
#W5A6 
def giai_thua(n):
    t= 1
    for i in range(1,n+1):
        t*=i
    return t 
n = int(input())
print(giai_thua(n))
```

```python
#W5A7 
def Cacl(a,b,op):
    if op=="+":
        return f"{a+b:.2f}" 
    if op=="-":
        return f"{a-b:.2f}"  
    if op=="/":
        return f"{a/b:.2f}"  
    if op=="*":
        return f"{a*b:.2f}"   
a,b = map(float,input().split())
op= input(" nhập vào phép toán cần tính") 
print(Cacl(a,b,op))
```

```python
#W5A8 
def hamming(a,b):
  n=a^b
  count=0
  while n>0:
    if n%2!=0:
      count+=1
    n//=2
  return count
x,y=map(int,input().split())
print(hamming(x,y))
```

```python
#W5A10 
def is_isomorphic(a, b):
    if len(a) != len(b):
        return False
    map_a = {}
    map_b = {}
    for i in range(len(a)):
        x = a[i]
        y = b[i]
        if x in map_a and map_a[x] != y:
            return False
        if y in map_b and map_b[y] != x:
            return False
        map_a[x] = y
        map_b[y] = x
    return True
a,b = map(str,input().split())
print(is_isomorphic(a,b))
```
