```python
#W13A1 
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        n = int(f.read().strip())
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev
if __name__ == "__main__":
    path = input("Nhap duong dan file: ")
    result = reverse(path)
    print(result)

```

```python
#W13A2 
import os
def canWinNim(path:str) ->bool:
    with open(path,'r',encoding='utf-8') as f:
        n=int(f.read().strip())
    return n%4!=0
if __name__=="__main__":
    filename="Win_Nim.txt"
    with open(filename,'w',encoding='utf-8') as f:
        f.write("4")
    if canWinNim(filename):
        print("WIN")
    else:
        print("LOSE")
```

```python
#W13A3 
def housesofhogwarts(path):
    file = open(path, "r")
    n = file.readlines().strip()
    file.close()
    for i in range(1, len(n)):
        print(n[i])
    file.close()    
path = "test.txt"
housesofhogwarts(path)        
```
