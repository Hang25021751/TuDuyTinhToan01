```python
#W7A1  
def t(arr,k):
    left, right=0 ,len(arr)-1 
    while left<=right: 
        mid= (left+right)//2 
        if arr[mid]==k: 
            return mid 
        elif arr[mid]>k: 
            right = mid -1 
        else: 
            left =mid+1 
    return - 1
arr= list(map(int,input().split()))
k = int(input())
print(t(arr,k))
```

```python
#W7A3
def bubble_sort(lst):
    n=len(lst)
    for i in range(n): 
        t=False
        for j in range(0,n-i-1): 
            if lst[j]>lst[j+1]:
                lst[j],lst[j+1]=lst[j+1],lst[j]
                t = True 
        if not t: 
            break 
    return lst 

arr=list(map(int,input().split()))
print(bubble_sort(arr))

def selection_sort(arr): 
    n =len(arr)
    for i in range(n):
        min_indx=i
        for j in range(i+1,n): 
            if arr[j]<arr[min_indx]:
                min_indx=j 
        arr[i],arr[min_indx]=arr[min_indx],arr[i]
    return arr

lst=list(map(int,input().split()))
print(selection_sort(lst))



```

```python
#W7A7 
def l(nums):
  if not nums:
    return 0
  n=len(nums)
  m = [1]*n
  for i in range(n):
    for j in range(i):
      if nums[j]<nums[i]:
        m[i]=max(m[i],m[j]+1)
  return max(m)
nums=eval(input())
print(l(nums))
```

```python
#W7A8 
def interval(intervals, queries):
    ans = []
    for i in queries:
        Min = 100000000000
        for left, right in intervals:
            if left <= i <= right:
                Min = min(Min, right - left + 1)
        if Min == 100000000000:
            ans.append(-1)
        else:
            ans.append(Min)
    return ans
nums = list(map(int, input().split()))
queries = list(map(int, input().split()))
intervals = []
tmp = []
for i in range(0, len(nums)):
    tmp.append(nums[i])
    if (i % 2 == 1):
        intervals.append(tmp)
        tmp = []
print(interval(intervals, queries))
```

```python
#W7A10 
import bisect 
def max_mod(nums,m):
  t1=0
  t2=0
  sorted_pre=[]
  for x in nums:
    t1=(t1+x)%m
    t2=max(t2,t1)
    id=bisect.bisect_right(sorted_pre,t1)
    if id<len(sorted_pre):
      t2=max(t2,(t1-sorted_pre[id]+m)%m) 
      bisect.insort(sorted_pre,t1) 
  return t2
num=eval(input())
m=int(input())
print(max_mod(num,m))
```

```python
#W7A13 
def h(a):
  dic={}
  for x in a:
    dic[x]=dic.get(x,0)+1
  sum=0
  for x in dic:
    if x+1 in dic:
      sum=max(sum,dic[x]+dic[x+1])
  return sum
a=list(map(int,input().split()))
print(h(a))
```
