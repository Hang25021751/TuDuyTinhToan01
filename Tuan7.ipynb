```python
#W7A1  
def t(arr,k):
    left, right=0 ,len(arr)-1 
    while left<=right: 
        mid= (left+right)//2 
        if arr[mid]==k: 
            return mid 
        elif arr[mid]>k: 
            right = mid -1 
        else: 
            left =mid+1 
    return - 1
arr= list(map(int,input().split()))
k = int(input())
print(t(arr,k))
```

```python
#W7A3
def bubble_sort(lst):
    n=len(lst)
    for i in range(n): 
        t=False
        for j in range(0,n-i-1): 
            if lst[j]>lst[j+1]:
                lst[j],lst[j+1]=lst[j+1],lst[j]
                t = True 
        if not t: 
            break 
    return lst 

arr=list(map(int,input().split()))
print(bubble_sort(arr))

def selection_sort(arr): 
    n =len(arr)
    for i in range(n):
        min_indx=i
        for j in range(i+1,n): 
            if arr[j]<arr[min_indx]:
                min_indx=j 
        arr[i],arr[min_indx]=arr[min_indx],arr[i]
    return arr

lst=list(map(int,input().split()))
print(selection_sort(lst))



```

```python
#W7A4
def so_lan(lst): 
    m=[]
    for i in lst: 
        b = lst.count(i) 
        m.append(b) 
    c = max(m) 
    for j in lst:
        if lst.count(j)==c:
            print(f"{j} xuat hien nhieu nhat,som nhat,{c} lan")
            break 

lst =list(map(int,input().split()))
so_lan(lst)



```
