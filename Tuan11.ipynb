```python
#W11A3 
try:
    a = int(input())
    if a <= 0:
        raise ValueError
    k= set()
    while a != 1:
        if a in k:
            print("NO")
            break
        k.add(a)
        s = 0
        for i in str(a):
            s += int(i) ** 2
        a = s
    if a == 1:
        print("YES")
except ValueError:
    print("ValueError")
```
