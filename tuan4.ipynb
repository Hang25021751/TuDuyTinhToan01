```python
#W4A6 in ra tổng các số nguyên tố trong đoạn a,b 
a,b = map(int,input("nhap vao 2 so nguyen duong:").split())
tong=0
for i in range(a,b+1):
    if i>1:
        t=True
        for j in range(2,int(i**0.5)+1):
            if i%j==0:
                t=False
                break 
    if t:
        tong+=i
print(tong)
    
```

```python
#W4A30 nhập vào một chuỗi đếm ra bao nhiêu kí tự chữ cái hoa , thường ,số 
n = input("nhap vao  mot chuoi:") 
t=0
m= 0 
z = 0 
for i in n:
    if 65<=ord(i)<=90:
        t+=1
    elif 97<=ord(i)<=122:
        m+=1
    elif i.isdigit():
        z+=1
print(f"so luong ki tu in hoa:{t}")
print(f"so luong ki tu  thuong:{m}")
print(f"so luong ki tu so :{z}")
```
